```python
import pandas as pd
```

```python
file_path = r"../files/sp24Enrollments.csv"
df = pd.read_csv(file_path, sep='|')
df
```

```python
df.keys()
```

```python
# combine rows based on same major
grouped_by_major = df.groupby('Major Name').agg({'College'      : 'first',
                                                 'Dept'         : 'first',
                                                 'Major Code'   : 'first',
                                                 'Total'     : 'sum'}).reset_index()

# formatting
grouped_by_major[['Dept', 'Major Code', 'Total']] = grouped_by_major[['Dept', 'Major Code', 'Total']].astype(int)
grouped_by_major['Major Name'] = grouped_by_major['Major Name'].str.strip()
grouped_by_major['College'] = grouped_by_major['College'].str.strip()

grouped_by_major
```

```python
grouped_by_major.to_csv('../sp24EnrollmentsClean.csv', sep='|', index=False)
```
